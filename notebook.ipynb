```python
data_dir = "/Users/melindans/VSCODE/E-Commerce-Data-Analysis-Project-on-macOs---DICODING/data/"

orders = pd.read_csv(data_dir + 'orders_dataset.csv')
items = pd.read_csv(data_dir + 'order_items_dataset.csv')
products = pd.read_csv(data_dir + 'products_dataset.csv')
payments = pd.read_csv(data_dir + 'order_payments_dataset.csv')
reviews = pd.read_csv(data_dir + 'order_reviews_dataset.csv')
customers = pd.read_csv(data_dir + 'customers_dataset.csv')
sellers = pd.read_csv(data_dir + 'sellers_dataset.csv')
geolocation = pd.read_csv(data_dir + 'geolocation_dataset.csv')
category = pd.read_csv(data_dir + 'product_category_name_translation.csv')

data = {'orders': orders,
        'items': items,
        'products': products,
        'payments': payments,
        'reviews': reviews,
        'customers': customers,
        'sellers': sellers,
        'geo': geolocation,
        'category': category}
```

### 1.b Assessing Data

#### 1.b.a Checking Data Types

```python
for df_name, df in data.items():
  print('\n', data[df_name].info())
```

```python
datetime_oo = ["order_purchase_timestamp","order_approved_at","order_delivered_carrier_date","order_delivered_customer_date","order_estimated_delivery_date"]
for column in datetime_oo:
  data['orders'][column] = pd.to_datetime(data['orders'][column])

datetime_oi = ["shipping_limit_date"]

for column in datetime_oi:
  data['items'][column] = pd.to_datetime(data['items'][column])

datetime_or = ["review_creation_date","review_answer_timestamp"]

for column in datetime_or:
  data['reviews'][column] = pd.to_datetime(data['reviews'][column])
```

#### 1.b.b Checking for Duplicates

```python
duplicate_counts = {}
total_sample_counts = {}

for df_name, df in data.items():
    duplicate_counts[df_name] = df.duplicated().sum()
    total_sample_counts[df_name] = len(df)

duplicate_counts_df = pd.DataFrame.from_dict(duplicate_counts, orient='index', columns=['Duplicate Count'])
total_sample_counts_df = pd.DataFrame.from_dict(total_sample_counts, orient='index', columns=['Total Sample Count'])

pd.concat([total_sample_counts_df, duplicate_counts_df], axis=1)
```

#### 1.b.c Checking for Missing Values

```python
null_counts = {}

for df_name, df in data.items():
    null_counts[df_name] = df.isna().sum()
    print(df_name, '\n', null_counts[df_name], '\n')
```

### 1.c Cleaning Data

#### 1.c.a Orders

```python
print(data['orders']['order_status'].value_counts())
```

```python
print('Taking only delivered orders we still get ' +
      str(round((data['orders'].order_status.value_counts()['delivered']/len(data['orders']))*100,2)) +
      '% of data.')
```

```python
delivered = data['orders']['order_id'][data['orders']['order_status'] != 'delivered'].values

for key, value in data.items():
    if 'order_id' in value.columns:
        value.drop(value.loc[value['order_id'].isin(delivered)].index, inplace = True)
```

```python
for key, value in data.items():
    for col in value.columns:
        if value[col].isnull().sum():
            print(col + ' (%): ' + str((sum(value[col].isnull())/len(value))*100))
```

```python
print('There are only ' + str(data['orders'].order_delivered_customer_date.isnull().sum()) +
      ' orders that are missing delivery dates.')
```

```python
null_deliveries = data['orders']['order_id'][data['orders']['order_delivered_customer_date'].isnull()].values

for key, value in data.items():
    if 'order_id' in value.columns:
        value.drop(value.loc[value['order_id'].isin(null_deliveries)].index, inplace = True)
```

```python
data['orders']['delivered_on_time'] = np.where(data['orders']['order_delivered_customer_date'] < data['orders']['order_estimated_delivery_date'], 'On Time', 'Late')
```

#### 1.c.b Reviews

```python
review_drop_cols = ['review_comment_title']
for col in review_drop_cols:
    if col in data['reviews'].columns:
        data['reviews'].drop(col, axis=1, inplace=True)
    else:
        print(f"Column '{col}' not found in DataFrame.")
```

```python
data['reviews']['review_comment_message'] = np.where(data['reviews']['review_comment_message'].isnull(), 0, 1)
```

#### 1.c.c Products

```python
product_drop_cols = ['product_name_length',
                     'product_description_length',
                     'product_weight_g',
                     'product_length_cm',
                     'product_height_cm',
                     'product_width_cm']

existing_cols_to_drop = [col for col in product_drop_cols if col in data['products'].columns]
data['products'].drop(existing_cols_to_drop, axis=1, inplace=True)

print('Product cols: ' , data['products'].columns.values)
```

```python
data['products'].loc[:, 'product_category_name'] = data['products']['product_category_name'].fillna(value='outro')
```

```python
data['products'].loc[:, 'product_photos_qty'] = data['products']['product_photos_qty'].fillna(value=0)
```

#### 1.c.d Geolocation

```python
data['geo'].drop_duplicates(inplace=True)
```

## **3. Visualization & Explanatory Analysis**

### 3.a How do daily orders and revenue fluctuate over time?


```python
order_data.sort_values(by="order_approved_at", inplace=True)
order_data.reset_index(drop=True, inplace=True)
daily_orders = order_data.groupby(order_data["order_approved_at"].dt.date).agg({
    "order_id": "count",
    "payment_value": "sum"
}).rename(columns={"order_id": "order_count", "payment_value": "revenue"}).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_orders, x="order_approved_at", y="order_count", marker="o", linewidth=2, color="#F47C20")
plt.title("Daily Orders Delivered", fontsize=22)
plt.xlabel("Date", fontsize=18)
plt.ylabel("Order Count", fontsize=18)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

The chart shows a steady upward trend in daily orders from 2016 to the end of 2017, with some significant spikes during the period. However, after reaching a peak, orders fluctuated and declined slightly at the end of 2018.

### 3.b Which products are the best sellers, and which ones are lagging in sales?


```python
item_sales = order_data.groupby("product_category_name_english").agg({
    "order_id": "count"
}).rename(columns={"order_id": "product_count"}).reset_index().sort_values(by="product_count", ascending=False)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x="product_count", y="product_category_name_english", data=item_sales.head(5), hue="product_category_name_english", dodge=False, palette="Blues_d")
plt.title("Top Sold Products", fontsize=22)
plt.xlabel("Sales Count", fontsize=18)
plt.ylabel(None)
plt.tight_layout()
plt.show()
```

The bed_bath_table category is the best-selling, followed by health_beauty, sports_leisure, computers_accessories, and furniture_decor. These products have a high number of sales, indicating great consumer demand in these categories.

```python
plt.figure(figsize=(8, 6))
sns.barplot(x="product_count", y="product_category_name_english", data=item_sales.tail(5), hue="product_category_name_english", dodge=False, palette="Reds_d")
plt.title("Least Sold Products", fontsize=22)
plt.xlabel("Sales Count", fontsize=18)
plt.ylabel(None)
plt.tight_layout()
plt.show()
```

The categories arts_and_craftmanship, la_cuisine, cds_dvds_musicals, fashion_childrens_clothes, and security_and_services are the least sold, with a much lower number of sales. This indicates that products in these categories are less desirable to consumers.
